```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def getimfolder(experiment):
    import os
    basedir = ""
    data = "E:\Raphael\Data"
    for folder in os.listdir(data):
        if str(experiment) in folder:
            basedir = os.path.join(data, folder)
    return basedir

experiment = 3
basedir = getimfolder(experiment)
print(basedir)
```

```python
df = pd.read_csv(os.path.join(basedir,"NNsimpleTest.csv"))
df.head()
```

```python


import seaborn as sns
sns.relplot(data=df, x="t", y="Ct", kind="line", hue="sigma",palette=sns.color_palette("rocket", as_cmap=True))
plt.show()
```
