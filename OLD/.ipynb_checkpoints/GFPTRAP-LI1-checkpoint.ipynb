```python
import pandas as pd
from tkinter import filedialog as fd
import os
```

```python
directory = fd.askdirectory()
dfa = []
```

```python
for filename in os.listdir(directory):
    location = directory + "/" + filename
    tdf = pd.read_csv(location)
    filename = filename[:len(filename)-4]
    rowcount = tdf.shape[0]
    group = []
    for i in range(rowcount):
        group.append(filename) 
    tdf['Group'] = group
    
    dfa.append(tdf)
print(dfa)
```

```python
df = pd.concat(dfa, ignore_index=True)


print(df)
```

```python
df.boxplot(by ='Group', column =['Mean'], grid = False)
```

```python
df.boxplot(by ='Group', column =['Median'], grid = False)
```

```python
df.boxplot(by ='Group', column =['Area'], grid = False)
```

```python
import scipy.stats as stats
```

```python
stats.f_oneway(df['Median'][df['Group'] == 'X'],
               df['Median'][df['Group'] == 'TP'],
               df['Median'][df['Group'] == 'FY'])
```

```python
import statsmodels.stats.multicomp as mc
```

```python

comp = mc.MultiComparison(df['Median'], df['Group'])
post_hoc_res = comp.tukeyhsd()
post_hoc_res.summary()
```

```python
stats.f_oneway(df['Area'][df['Group'] == 'X'],
               df['Area'][df['Group'] == 'TP'],
               df['Area'][df['Group'] == 'FY'])
```

```python
df.boxplot(by ='Group', column =['Min'], grid = False)
```

```python
df.boxplot(by ='Group', column =['Max'], grid = False)
```

```python
df.sort_values(by='Group', ascending=False)
print(df)
```

```python
from scipy.stats import kstest, mannwhitneyu
kres = kstest(df['Median'], 'norm')
sns.histplot(data=df, x="Median")
n = kres[1]
print(n)
if n < 0.05:
    stat_results = [
      mannwhitneyu(df['Median'][df['Group'] == 'X'], df['Median'][df['Group'] == 'TP'], alternative="two-sided"),
      mannwhitneyu(df['Median'][df['Group'] == 'TP'], df['Median'][df['Group'] == 'FY'], alternative="two-sided"),
      mannwhitneyu(df['Median'][df['Group'] == 'X'], df['Median'][df['Group'] == 'FY'], alternative="two-sided")
        ]

               
    pvalues = [result.pvalue for result in stat_results]

    print("X vs TP: \n", stat_results[0], "\n")
    print("TP vs FY: \n", stat_results[1], "\n")
    print("X vs FY: \n", stat_results[2], "\n")
    print(pvalues)
else:
    comp = mc.MultiComparison(df['Median'], df['Group'])
    post_hoc_res = comp.tukeyhsd()
    post_hoc_res.summary()
```
