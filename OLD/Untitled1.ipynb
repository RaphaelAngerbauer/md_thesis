```python
import numpy as np
import seaborn as sns
from scipy.stats import binom
```

```python
true_Ct = 0.01
true_Co = 0.001
N1 = 10000
N2 = 1000000
Ct = binom.rvs(N1, true_Ct, size=1000000)/N1
Co = binom.rvs(N2, true_Co, size=1000000)/N2
Ct[Ct == 0] = 0.00000001
Co[Co == 0] = 0.00000001
Ct_rel = Ct/(1-Ct)
Co_rel = Co/(1-Co)
Ct_log = np.log(Ct_rel)
Co_log = np.log(Co_rel)
E = Ct_log-Co_log
sns.kdeplot(data=E)
```
