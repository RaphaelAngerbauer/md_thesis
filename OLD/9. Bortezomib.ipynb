```python
import pandas as pd
import pathophys as pp
spec = "9"
df = pp.get_df_flourometry(spec, True)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
print(df["Mean_norm_X"])
```

```python
df["Group2"] = df["Group2"].replace("Null",0)
df.head()
```

```python
loc = os.path.join("D:\Raphael\Data\9. Bortezomib\Flourometry","ResultsBort.xlsx")
df2 = pd.read_excel(loc)
df = pd.concat([df,df2])
df.head()
```

```python
dfs = []
for t in df["Try"].unique():
    dfg1 = df[df["Try"]==t]
    for g2 in dfg1["Group2"].unique():
        dfg2 = dfg1[dfg1["Group2"]==g2]
        m = dfg2["Mean_sub"][dfg2["Group1"]=="X"].mean()
        dfg2["Mean_norm"] =  dfg2["Mean_sub"]/m
        dfs.append(dfg2)
df = pd.concat(dfs, ignore_index = True)
df.head()
```

```python
params = pp.loadparams()
df["Group2"] = df["Group2"].astype(int)
df = df.drop(df[df["Group2"]==160].index)
cc = 0

ax = plt.subplot(111)
for g1 in df["Group1"].unique():
    gdf = df[df["Group1"]==g1]
    X = list(gdf.groupby("Group2").groups.keys())
    Y = list(gdf.groupby("Group2")["Mean_norm"].mean().sort_index().values)
    print(Y)
    std = gdf.groupby("Group2").std().sort_index()["Mean_norm"].values
    count = gdf.groupby("Group2").count().sort_index()["Mean_norm"].values
    CI = list(1.96*(std)/np.sqrt(count))
    g = g1
    if g1 == "X":
        g = "Ctrl"
    ax.plot(X,Y,params["Colors"][cc], label = g,linewidth=2, marker = "d")
    
    plt.errorbar(X,Y,yerr=CI,capsize=4,elinewidth=0.5,linestyle="",color = params["Colors"][cc])
    cc = cc+1
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
    if axis in ['top','right']:
        ax.spines[axis].set_visible(False)
ax.legend(loc=3)
ax.set_xlabel('[Bortezomib](nM)', fontsize=11)
ax.set_ylabel('Rel. Fluorescence', fontsize=11)
ax.set_title('Fluorometry (normalized)', fontsize=11)
ax.set_title("a", loc="left", fontsize=20, fontweight="bold", pad=15)
ax.tick_params(axis='both', which='major', width = 2)
plt.show()
```

```python
import scipy.stats as stats
for g2 in df["Group2"].unique():
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
    gdf = df[df["Group2"]==g2]
    fvalue, pvalue = stats.f_oneway(gdf[gdf["Group1"]=="X"]["Mean_norm"], gdf[gdf["Group1"]=="TP"]["Mean_norm"], gdf[gdf["Group1"]=="FY"]["Mean_norm"])
    print(g2)
    print(fvalue, pvalue)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
for g2 in df["Group2"].unique():
    gdf = df[df["Group2"]==g2]
    print(g2)
    m_comp = pairwise_tukeyhsd(endog=gdf['Mean_norm'].dropna(), groups=gdf['Group1'], alpha=0.05)
    print(m_comp.summary())
```
